```python
import pandas as pd
import numpy as np
import seaborn as sns
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import nltk
from nltk import tokenize
from nltk.probability import FreqDist
import urllib.request
nltk.download('punkt')
%matplotlib inline
```

```python
textfile = open('C:/Users/user/Desktop/final/Agatha Christie - Death On The Nile.txt', 'r').read()
print(textfile[1000:1500])
```

```python
from nltk import word_tokenize
words = word_tokenize(textfile)
words
print(f"Total number of words in the file is: {len(words)}")
```

```python
fdist = FreqDist(words)
fdist.most_common(10)
```

```python
words_no_punc = []
for word in words:
    if word.isalpha():
        words_no_punc.append(word.lower())
print(f"Words without punciation is: {len(words_no_punc)}")
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
stopwords_list = stopwords.words("english")
print(stopwords_list)
```

```python
clean_words = []
for word in words_no_punc:
    if word not in stopwords_list:
        clean_words.append(word)
print(f"Words without punication and stopwords is: {len(clean_words)}")
```

```python
clean_words_string = " ".join(clean_words)
wordcloud = WordCloud(background_color = 'white').generate(clean_words_string)
plt.figure(figsize = (7,7))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
